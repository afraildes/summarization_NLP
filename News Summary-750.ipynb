```python
import torch
from transformers import BertTokenizerFast, EncoderDecoderModel
```

```python
ckpt = 'mrm8488/bert2bert_shared-turkish-summarization'
tokenizer = BertTokenizerFast.from_pretrained(ckpt)
model = EncoderDecoderModel.from_pretrained(ckpt)

def generate_summary(text):

   inputs = tokenizer([text], padding="max_length", truncation=True, max_length=512, return_tensors="pt")
   input_ids = inputs.input_ids
   attention_mask = inputs.attention_mask
   output = model.generate(input_ids, attention_mask=attention_mask)
   return tokenizer.decode(output[0], skip_special_tokens=True)
   
text = "Öğrenci Seçme ve Yerleştirme Merkezi (ÖSYM), 12 Şubat'ta sona eren YGS' ye başvuru süresini 16 Şubat Salı günü saat 17.00'e kadar uzatmıştı. ÖSYM, başvuru süresinin tekrar uzatılmasının mümkün olmayacağını bildirdi ve 2010 ÖSYS kapsamında yapılacak YGS ve YLS sınavlarının daha önce açıklanan tarihlerde yapılacağını, sınav takviminde bir değişiklik olmayacağını da açıkladı. ÖSYM Başkanı Ünal Yarımağan ise, Danıştay'ın, YÖK'ün katsayı uygulamasının yürütmesini ikinci kez durdurmasının ardından adaylara seslenerek, Bunlar adayları tamamen etkilemez demek istemiyorum, ama çok etkilememesi gerekir. Adaylar bu konulardan kendilerini biraz soyutlayıp dersleriyle ilgilensinler, sınav hazırlıklarını sürdürsünler. Adaylardan bunu rica ettim. Başvuru süresini 2 gün uzattık. Sınava girmeyi düşünen tüm adaylar lütfen girsinler, başvurularını yapsınlar diye konuşmuştu."
generate_summary(text)
```

```python
summary_model = []
for i in range(len(text_all)):
    summary_model.append(generate_summary(text_all[i]).replace('"', ""))
```
